**파이썬으로 선형 회귀 적용하기 (경사하강법)**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

```python
plt.scatter(x, y)
plt.show()
```

```python
a = 0
b = 0

lr = 0.03
epochs = 2001
```

```python
n = len(x)

for i in range(epochs):
  y_pred = a*x+b
  error = y-y_pred
  
  a_diff = (2/n)*sum(-x*(error))
  b_diff = (2/n)*sum(-(error))

  a = a-lr*a_diff
  b = b-lr*b_diff

  if i%100 == 0:
    print(f"epoch={i}, 기울기={a}, 절편={b}")
```

```python
y_pred = a*x+b

plt.scatter(x,y)
plt.plot(x,y_pred,'r')
plt.show()
```

**파이썬으로 다중 선형 회귀 적용하기**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y)
plt.show()
```

```python
a1 = 0
a2 = 0
b = 0

lr = 0.01

epochs = 2001
```

```python
n = len(x)

for i in range(epochs):
  y_pred = a1*x1+a2*x2+b
  error = y-y_pred
  
  a1_diff = (2/n)*sum(-x1*(error))
  a2_diff = (2/n)*sum(-x2*(error))
  b_diff = (2/n)*sum(-(error))

  a1 = a1 - lr*a1_diff
  a2 = a2 - lr*a2_diff
  b = b - lr*b_diff

  if i%100==0:
    print(f"epoch={i}, 기울기1={a1}, 기울기2={a2}, 절편={b}")
```

```python
print("실제 점수:",y)
print("예측 점수:",y_pred)
```

**텐서플로에서 실행하는 선형 회귀 모델**

```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

```python
model = Sequential()

model.add(Dense(1, input_dim=1, activation = 'linear'))
model.compile(optimizer='sgd', loss='mse')
model.fit(x, y, epochs=2000)
```

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()
```

```python
hour = 7
prediction = model.predict([hour])

print(f"{hour}시간을 공부할 경우의 예상 점수는 {prediction}점입니다.")
```

**텐서플로에서 실행하는 다중 선형 회귀 모델**

```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([[2, 0],[4, 4],[6, 2], [8, 3]])
y = np.array([81, 93, 91, 97])
```

```python
model = Sequential()

model.add(Dense(1, input_dim=2, activation='linear'))
model.compile(optimizer='sgd', loss='mse')

model.fit(x, y, epochs=2000)
```

```python
hour = 7
private_class = 4

prediction = model.predict([[hour, private_class]])

print(f"{hour}시간을 공부하고 {private_class}시간의 과외를 받을 경우, 예상 점수는 {prediction}입니다.")
```
